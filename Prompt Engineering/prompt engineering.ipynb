```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
```

```python
model_name = "microsoft/phi-3-mini-4k-instruct"
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype="auto",
    trust_remote_code=True,

)
tokenizer = AutoTokenizer.from_pretrained(model_name)
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=True,
    top_p=0.8,
    temperature=1

)
```

```python
# prompt
messages = [{"role":"user","content":"create a funny joke about chickens"}]
output = pipe(messages,temperature=0.9,do_sample=True,top_p=0.3)
output[0]['generated_text']
```

```python
prompt=pipe.tokenizer.apply_chat_template(messages,tokenize=False)
print(prompt)
```

```python
messages = [{"role":"user","content":"return single next word for the statement: The sky is"}]
output1 = pipe(messages)
print(output1[0]['generated_text'])
```

```python
import PyPDF2

def read_pdf_content(filename):
    """Reads the text content from a PDF file.

    Args:
        filename: The path to the PDF file.

    Returns:
        A string containing the extracted text.
    """

    with open(filename, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text

# Example usage:
pdf_file_path = "/content/drive/MyDrive/GENAI_Work/Hands-on LLM/test_pdf.pdf"
pdf_text = read_pdf_content(pdf_file_path)
print(pdf_text)
```

```python
# components of prompt
persona = "You are an expert in large language models. You excel at breadking down complex papers into digestible summaries. \n"
instruction = "Summarize the key findings of the paper provided. \n"
context = "You summary should extract the most crucial points that can help researchers quickly understand the most vital information of the paper. \n"
data_format="Create a bullet point summary that outlines the method. Follow this up with a concise paragraph that encapsulates the main results. \n"
audience="The summary is designed for busy researchers that quickly need to grasp the newest trends in Large Language Models. \n"
tone="The tone should be professional and clear"
text=pdf_text
data = f"Paper to summarize:{text}"
# full prompt
prompt=persona+instruction+context+data_format+audience+tone+data
```

```python
messages5 = [{"role":"user","content":prompt}]
output5 = pipe(messages5)
print(output5[0]['generated_text'])
```
